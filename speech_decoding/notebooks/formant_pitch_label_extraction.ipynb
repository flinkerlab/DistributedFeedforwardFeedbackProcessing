# formant and pitch extraction function

```python
import numpy as np
import parselmouth
import librosa
from IPython.display import Audio
import IPython.display as ipd
from librosa import display as librosadisplay
%pylab inline

class PraatExtractor():
    def __init__(self, sampling_rate, frame_rate, wav_file_path = '', audio = None):
        self.sampling_rate = sampling_rate
        self.frame_rate = frame_rate
        if audio is not None:
            self.audio = audio
        else:
            self.audio, sr = librosa.core.load(file_path, sr = sampling_rate)            
        self.snd = parselmouth.Sound(audio, sampling_frequency = sampling_rate, start_time = 0)
        self.out_frames = int(audio.shape[0] / sampling_rate * frame_rate)
        
    def extract_formants(self, formants_num):
        #formant: https://www.fon.hum.uva.nl/praat/manual/Sound__To_Formant__burg____.html
        #note that formant frame rate is not accurate, interpolation is neeeded
        #should use f0 or other component to filter out the silent period, 
        #e.g.: formants_interp[self.pitch_interp == 0,:] = np.nan
        formants = self.snd.to_formant_burg(max_number_of_formants = formants_num, time_step = 1 / self.frame_rate, window_length = 0.025)
        formants_ts = parselmouth.praat.call(formants, "To Matrix", 1).values[0,:].reshape(-1, 1)

        for formant_idx in range(2, formants_num + 1):
            formants_ts = np.concatenate((formants_ts, parselmouth.praat.call(formants, "To Matrix", formant_idx).values[0,:].reshape(-1, 1) ), axis = 1)

        #interpolation
        if formants_ts.shape[0] != self.out_frames:
            self.formants_interp = np.zeros([self.out_frames, formants_num])
            for n_formant in range(formants_num):
                self.formants_interp[:, n_formant] = np.interp(range(self.out_frames), range(formants_ts[:, n_formant].shape[0]), formants_ts[:, n_formant])
        
        return self.formants_interp
    
    def extract_pitch(self,):
        self.pitch = self.snd.to_pitch(time_step = 1 / self.frame_rate).selected_array['frequency']
        self.pitch_interp = np.interp(range(self.out_frames), range(self.pitch.shape[0]), self.pitch)
        return self.pitch_interp

    def extract_intensity(self):
        intensity = self.snd.to_intensity().values.ravel()
        if intensity.shape[0] != self.out_frames:
            self.intensity_interp = np.interp(range(self.out_frames), range(intensity.shape[0]), intensity)
        return self.intensity_interp
            
    def draw_components(self, n_fft = 256, play_sound = True, plot_intensity = False):
        if play_sound:
            display(ipd.Audio(self.audio, rate = self.sampling_rate))
        spectrogram = librosa.stft(self.audio, n_fft = n_fft * 2 - 1, win_length = n_fft * 2 - 1, \
                               hop_length = int(self.sampling_rate / self.frame_rate))#n_fft//4)#int(self.sampling_rate / self.frame_rate))
        #print (spectrogram.shape)
        fig, ax = plt.subplots(figsize = (spectrogram.shape[1] / spectrogram.shape[0] * 8, 8))
        librosa.display.specshow(librosa.amplitude_to_db(abs(spectrogram)), cmap = mpl.cm.gray_r,#cm.Blues
                                              sr = sampling_rate, ax = ax, y_axis = 'linear')
        ax.plot(self.pitch_interp, 'o', label = 'f0', linewidth = 3,  markersize = 4, color='w')
        self.formants_interp[self.pitch_interp == 0,:] = np.nan
        for formant_idx in range(self.formants_interp.shape[1]):
            ax.plot(self.formants_interp[:,formant_idx], 'o', linewidth = 2,  markersize = 5, label = 'f{}'.format(formant_idx + 1))
        ax.legend(loc = 1)
        ax.set_ylabel("frequency [Hz]")
        
        if plot_intensity:
            ax2 = ax.twinx()
            ax2.plot(self.intensity_interp, color = 'w', linewidth = 3)
            ax2.plot(self.intensity_interp, label = 'intensity', linewidth = 1)
            ax2.legend(loc = 2)
            ax2.set_ylabel("intensity [dB]")
        
        
        
class PraatExtractorRefiner(PraatExtractor):
    def __init__(self, fills = 3, **kwargs):
        super().__init__(**kwargs)
        self.fills = fills
        
    def arr_non_decrease_modifier(self, arr_inp, increased_value = 1000): 
        ind = np.where(arr_inp[1:] - arr_inp[:-1] <= 0)[0]
        arr_inp[ind + 1] = arr_inp[ind] + increased_value
        return arr_inp
    
    def find_nearest(self, array, value): 
        return array[(np.abs(array - value)).argmin()]
 
    def refine_formants(self, formants_num):
        # to refine formants, within each time step:
        # make sure non-decreasing in freq dim, modify the freq by t-1, use arr_non_decrease_modifier
        # make sure every start_ind's freq is non-decreasing, then fix f_t by finding nearest neighbor in f_t-1, then make sure it is non-decreasing again
        if not hasattr(PraatExtractorRefiner, 'pitch_interp'):
            self.pitch_interp = self.extract_pitch()
        if not hasattr(PraatExtractorRefiner, 'formants_interp'):
            self.formants_interp = self.extract_formants(formants_num)
        
        silent_region_index = np.zeros_like(self.pitch_interp)
        silent_region_index[self.pitch_interp > 0] = 1
        silent_region_ind = np.where(silent_region_index[1:] - silent_region_index[:-1] != 0)[0]
        if silent_region_index[0] > 0:
            silent_region_ind = np.concatenate((np.array([0]), silent_region_ind), axis = 0)
        if silent_region_index[-1] > 0:
            silent_region_ind = np.concatenate((silent_region_ind, np.array([self.pitch_interp.shape[0]])), axis = 0)
        for start_ind, end_ind in silent_region_ind.reshape(-1, 2):
            self.formants_interp[start_ind] = self.arr_non_decrease_modifier(self.formants_interp[start_ind])
            for loc_ind in range(start_ind + 1, end_ind):
                for formant_idx in range(self.formants_interp.shape[1]):
                    self.formants_interp[loc_ind, formant_idx] = self.find_nearest(self.formants_interp[loc_ind], self.formants_interp[loc_ind - 1, formant_idx])
                self.formants_interp[loc_ind] = self.arr_non_decrease_modifier(self.formants_interp[loc_ind])
        #deal with shifting issues
        fill_arr = np.zeros([self.fills, self.formants_interp.shape[1]])
        fill_arr[:] = np.nan
        self.formants_interp = np.concatenate((fill_arr, self.formants_interp), axis = 0)[:-self.fills]
        return self.formants_interp
```
